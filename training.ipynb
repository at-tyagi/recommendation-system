```python
def prep_dict_ro_list(genres):
    L=[]
    genres=ast.literal_eval(genres)
    for i in genres:
        L.append(i["name"])
    return L
```

```python
def norm(column):
    ans=column-column.mean()
    ans=ans/column.std()
    return round(ans,2)
```

```python
df2['budget']=norm(df2['budget'])
df2['budget']=df2['budget'].apply(lambda x: [x])
df2['popularity']=norm(df2['popularity'])
df2['popularity']=df2['popularity'].apply(lambda x: [x])
df2['revenue']=norm(df2['revenue'])
df2['revenue']=df2['revenue'].apply(lambda x: [x])
df2['vote_average']=norm(df2['vote_average'])
df2['vote_average']=df2['vote_average'].apply(lambda x: [x])
df2['vote_count']=norm(df2['vote_count'])
df2['vote_count']=df2['vote_count'].apply(lambda x: [x])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
mat=cv.fit_transform(df4['tags']).toarray()
```

```python
mat.shape
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()                                ##### convert loved, loving to love ,, danceing dance to danc
```

```python
def stem(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    l=" ".join(l)
    return l
```

```python
# finding cosine distance b/w movie vectors
# cosine similarity = 1/ cosine distance =1/ angle bw vectors
```

```python
from sklearn.metrics.pairwise  import cosine_similarity
```

```python
similarity=cosine_similarity(mat)
```

```python
def recommend(movie):
    try:
        indx=df4[df4['title']==movie.lower()].index[0]
        movie_list=sorted(list(enumerate(similarity[indx])), reverse=True, key=lambda x: x[1])[1:6]
        for i,_ in movie_list:
            print(df4.iloc[i]['title'])
    except IndexError:
        print('Movie Not Found !!!')
```
